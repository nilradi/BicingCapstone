```python
top_10_stations = df_2023.groupby('station_id')['capacity'].max().nlargest(10).index
top_stations_data = df_2023[df_2023['station_id'].isin(top_10_stations)]
top_stations_data['datetime'] = pd.to_datetime(top_stations_data['datetime'])

top_stations_data['date'] = top_stations_data['datetime'].dt.date
daily_data = top_stations_data.groupby(['station_id', 'date']).agg({
    'num_docks_available': 'mean',
    'capacity': 'first'
}).reset_index()

for station_id in top_10_stations:
    station_data = daily_data[daily_data['station_id'] == station_id]

    plt.figure(figsize=(14, 8))
    plt.plot(station_data['date'], station_data['num_docks_available'], label='Docks Available')
    plt.plot(station_data['date'], station_data['capacity'], linestyle='--', label='Capacity')

    plt.xlabel('Date')
    plt.ylabel('Number of Docks')
    plt.title(f'Station {station_id} - Docks Available and Capacity Over Time in 2023')
    plt.legend()
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
df_2023 = df[df['year'] == 2023]

# Convertir la columna 'datetime' a tipo datetime
df_2023['datetime'] = pd.to_datetime(df_2023['datetime'])

# Agrupar por da y calcular la media de los muelles disponibles
df_2023['date'] = df_2023['datetime'].dt.date
daily_data = df_2023.groupby(['station_id', 'date']).agg({
    'num_docks_available': 'mean'
}).reset_index()

# Convertir la columna 'date' a tipo datetime
daily_data['date'] = pd.to_datetime(daily_data['date'])

# Filtrar datos para enero y diciembre de 2023
january_data = daily_data[daily_data['date'].between(pd.to_datetime('2023-01-01'), pd.to_datetime('2023-01-31'))]
december_data = daily_data[daily_data['date'].between(pd.to_datetime('2023-12-01'), pd.to_datetime('2023-12-31'))]

# Calcular la media de docks available por estacin para enero y diciembre
january_mean = january_data.groupby('station_id')['num_docks_available'].mean().reset_index(name='january_mean')
december_mean = december_data.groupby('station_id')['num_docks_available'].mean().reset_index(name='december_mean')

# Fusionar las medias de enero y diciembre
growth_data = pd.merge(january_mean, december_mean, on='station_id')

# Calcular el crecimiento
growth_data['growth'] = growth_data['january_mean'] - growth_data['december_mean']

# Merge de nuevo con el DataFrame original
df_2023 = pd.merge(df_2023, growth_data[['station_id', 'growth']], on='station_id', how='left')

# Seleccionar las 15 estaciones con mayor crecimiento
top_15_growth_stations = growth_data.nlargest(15, 'growth')

# Obtener la lista de station_id
top_15_station_ids = top_15_growth_stations['station_id'].tolist()

# Filtrar datos para estas 15 estaciones
top_15_stations_data = df_2023[df_2023['station_id'].isin(top_15_station_ids)]
```

```python
import plotly.express as px

# IDs con ms desviacin en predicciones
ids_mas_desviacion = [230, 153, 465, 32, 42, 390, 280, 73, 33, 12, 112, 40, 65, 446, 364]

# IDs con menos desviacin en predicciones
ids_menos_desviacion = [474, 258, 483, 271, 301, 484, 490, 268, 463, 518, 491, 270, 10, 269, 473]

# Combinar ambas listas de IDs
ids_combinados = ids_mas_desviacion + ids_menos_desviacion

# Filtrar el DataFrame
filtered_data = map_data.loc[map_data.index.isin(ids_combinados)].copy()

# Crear una lista de categoras basado en los IDs filtrados
categories = []
for id in filtered_data.index:
    if id in ids_mas_desviacion:
        categories.append('Ms Desviacin')
    else:
        categories.append('Menos Desviacin')

# Aadir la columna de categora al DataFrame filtrado
filtered_data['category'] = categories

# Crear el mapa
fig = px.scatter_mapbox(
    filtered_data,
    lat="lat",
    lon="lon",
    size="capacity",
    hover_name="address",  # Aadir direccin a la informacin de hover
    color="category",  # Usar la columna de categora para diferenciar los puntos
    color_discrete_map={'Ms Desviacin': 'red', 'Menos Desviacin': 'blue'},  # Colores para cada categora
    size_max=15,
    zoom=10,
    mapbox_style="carto-positron",
    title="Bicing Stations Altitude and Capacity in 2024",
    width=1200, height=900
)

# Actualizar el layout para mejor visualizacin
fig.update_layout(
    autosize=True,
    hovermode="closest",
    title={
        'text': "Bicing Stations with Higher and Lower Deviation in Predictions",
        'y':0.9,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'},
    legend_title="Categora",
    coloraxis_showscale=False  # Ocultar la barra de color ya que usamos colores discretos
)

# Mostrar la figura
fig.show()
```
